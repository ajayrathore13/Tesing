```python
import pandas as pd
import os
import openpyxl
import pyxlsb
import schedule
import time
import datetime
def PastMonths():
    if datetime.date.today().day <= 3:
    # Define the path to the folder containing the Excel files
        folder_path = r"//INN5FILER1V0001.corp.idemia.com/Work/Banking Bureau/Data/BPB MIS/DAILY BPB MIS/1. DPR DATA/1. 2023"

        # Create an empty list to store the Excel files
        excel_files = []

        # Loop through the directory and its sub-directories to find Excel files
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                if file.endswith(".xlsb")  and "DPR" in file:
                    excel_files.append(os.path.join(root, file))

        # Define the name of the worksheet you want to combine
        sheet_name = "DPR "

        # Define the list of columns you want to include
        column_list = ["PR No", "Perso Date", "Perso Qty.","PR Qty", "Total Rej. Qty", "After perso pending for FF",
                    "Machine Name","Bank"]

        # Create an empty DataFrame to store the combined data
        combined_df = pd.DataFrame()

        # Loop through each Excel file and read the specified worksheet into a DataFrame
        for file in excel_files:
            try:
                df = pd.read_excel(file, sheet_name=sheet_name, engine='pyxlsb', header=5)
                df = df[column_list] # select only specified column 
                df['Perso Date'] = pd.to_datetime(df['Perso Date'] , unit='D', origin='1899-12-30')
                #df['Perso Date'] = pd.to_datetime(df['Perso Date'])
                df['Year'] = df['Perso Date'].dt.year
                df['Month'] = df['Perso Date'].dt.month
                max_date = df['Perso Date'].max()
                max_year = max_date.year
                max_month = max_date.month
                df = df.loc[(df['Year'] == max_year) & (df['Month'] == max_month)]
                df = df[df['Perso Date'].notna()]
                combined_df = combined_df.append(df, ignore_index=True)

            except Exception as e:
                print(f"Error reading worksheet '{sheet_name}' from file '{file}': {e}")

            # specify the file name
        file_name = 'r"//INN5FILER1V0001.corp.idemia.com/Work/FI Data Analytics/02. Datasets/DPR/PastMonths.xlsx"'     
            # check if the file already exists
        if os.path.exists(file_name):
            os.remove(file_name)  # delete the file if it exists

        # Write the combined data to a new Excel file
        #combined_df

        combined_df.to_excel(r"//INN5FILER1V0001.corp.idemia.com/Work/FI Data Analytics/02. Datasets/DPR/PastMonths.xlsx", index=False)
        
def CurrentMonth():
    # Define the path to the folder containing the Excel files
    folder_path = r"//INN5FILER1V0001.corp.idemia.com/Work/Banking Bureau/Data/BPB MIS/DAILY BPB MIS/NOIDA DAILY BPB MIS"

    # Create an empty list to store the Excel files
    excel_files = []

    # Loop through the directory and its sub-directories to find Excel files
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".xlsb")  and "DPR" in file:
                excel_files.append(os.path.join(root, file))

    # Define the name of the worksheet you want to combine
    sheet_name = "DPR "

    # Define the list of columns you want to include
    column_list = ["PR No", "Perso Date", "Perso Qty.","PR Qty", "Total Rej. Qty", 
                   "After perso pending for FF","Machine Name","Bank"]

    # Create an empty DataFrame to store the combined data
    combined_df = pd.DataFrame()

    # Loop through each Excel file and read the specified worksheet into a DataFrame
    for file in excel_files:
        try:
            df = pd.read_excel(file, sheet_name=sheet_name, engine='pyxlsb', header=5)
            df = df[column_list] # select only specified column 
            df['Perso Date'] = pd.to_datetime(df['Perso Date'] , unit='D', origin='1899-12-30')
    #         df['Perso Date'] = pd.to_datetime(df['Perso Date'])
            df['Year'] = df['Perso Date'].dt.year
            df['Month'] = df['Perso Date'].dt.month
            max_date = df['Perso Date'].max()
            max_year = max_date.year
            max_month = max_date.month
            df = df.loc[(df['Year'] == max_year) & (df['Month'] == max_month)]
            df = df[df['Perso Date'].notna()]
            combined_df = combined_df.append(df, ignore_index=True)
            
        except Exception as e:
            print(f"Error reading worksheet '{sheet_name}' from file '{file}': {e}")
                    # specify the file name
    file_name = 'r"//INN5FILER1V0001.corp.idemia.com/Work/FI Data Analytics/02. Datasets/DPR/CurrentMonth.xlsx"'     
            # check if the file already exists
    if os.path.exists(file_name):
        os.remove(file_name)  # delete the file if it exists

    # Write the combined data to a new Excel file
    combined_df.to_excel(r"//INN5FILER1V0001.corp.idemia.com/Work/FI Data Analytics/02. Datasets/DPR/CurrentMonth.xlsx", index=False)

schedule.every().day.at("16:55:40").do(PastMonths)
schedule.every().day.at("14:30").do(CurrentMonth) 
schedule.every().day.at("16:55").do(CurrentMonth)  
#schedule.every().day.at("17:12").do(PastMonths) 
while True:
    schedule.run_pending()
    time.sleep(1)

```
