```python
import pandas as pd
import os
import openpyxl
import pyxlsb
import schedule
import time
import datetime

def PPNoidaPM():
    if datetime.date.today().day <= 3:
        # Define the path to the folder containing the Excel files
        folder_path = r"//INN5FILER1V0001.corp.idemia.com/Work/Banking Bureau/Data/BPB MIS/DAILY BPB MIS/1. DPR DATA/1. 2023"

        # Create an empty list to store the Excel files
        excel_files = []

        # Loop through the directory and its sub-directories to find Excel files
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                if file.endswith(".xlsb")  and "DPR" in file:
                    excel_files.append(os.path.join(root, file))

        # Define the name of the worksheet you want to combine
        sheet_name = "Banking Perso performance"


        # Define the list of columns you want to include
        column_list = ["Date","Total Man Power (as per HR Record)","Total Present",
                       "OT Operators","Total OT HRs","Total Man Hours"]

        # Create an empty DataFrame to store the combined data
        combined_df = pd.DataFrame()

        # Loop through each Excel file and read the specified worksheet into a DataFrame
        for file in excel_files:
            try:
                df = pd.read_excel(file, sheet_name=sheet_name, engine='pyxlsb', header=1)
                df = df[column_list] # select only specified column 
                df['Date'] = pd.to_datetime(df['Date'] , unit='D', origin='1899-12-30')
                df = df[df['Date'].notna()]
                combined_df = combined_df.append(df, ignore_index=True)

            except Exception as e:
                print(f"Error reading worksheet '{sheet_name}' from file '{file}': {e}")

            # specify the file name
        file_name = 'r"//INN5FILER1V0001.corp.idemia.com/Work/FI Data Analytics/02. Datasets/DPR/PP_NoidaPM.xlsx"'     
            # check if the file already exists
        if os.path.exists(file_name):
            os.remove(file_name)  # delete the file if it exists

        # Write the combined data to a new Excel file
        #combined_df

        combined_df.to_excel(r"//INN5FILER1V0001.corp.idemia.com/Work/FI Data Analytics/02. Datasets/DPR/PP_NoidaPM.xlsx", index=False)
        
def PPNoidaCM():
    # Define the path to the folder containing the Excel files
    folder_path = r"//INN5FILER1V0001.corp.idemia.com/Work/Banking Bureau/Data/BPB MIS/DAILY BPB MIS/NOIDA DAILY BPB MIS"

    # Create an empty list to store the Excel files
    excel_files = []

    # Loop through the directory and its sub-directories to find Excel files
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".xlsb")  and "DPR" in file:
                excel_files.append(os.path.join(root, file))

    # Define the name of the worksheet you want to combine
    sheet_name = "Banking Perso performance"


    # Define the list of columns you want to include
    column_list = ["Date","Total Man Power (as per HR Record)","Total Present",
                   "OT Operators","Total OT HRs","Total Man Hours"]

    # Create an empty DataFrame to store the combined data
    combined_df = pd.DataFrame()

    # Loop through each Excel file and read the specified worksheet into a DataFrame
    for file in excel_files:
        try:
            df = pd.read_excel(file, sheet_name=sheet_name, engine='pyxlsb', header=1)
            df = df[column_list] # select only specified column 
            df['Date'] = pd.to_datetime(df['Date'] , unit='D', origin='1899-12-30')
            df = df[df['Date'].notna()]
            combined_df = combined_df.append(df, ignore_index=True)

        except Exception as e:
            print(f"Error reading worksheet '{sheet_name}' from file '{file}': {e}")

        # specify the file name
    file_name = 'r"//INN5FILER1V0001.corp.idemia.com/Work/FI Data Analytics/02. Datasets/DPR/PP_NoidaCM.xlsx"'     
        # check if the file already exists
    if os.path.exists(file_name):
        os.remove(file_name)  # delete the file if it exists

    combined_df.to_excel(r"//INN5FILER1V0001.corp.idemia.com/Work/FI Data Analytics/02. Datasets/DPR/PP_NoidaCM.xlsx", index=False)

def PP_MumbaiPM():
    if datetime.date.today().day <= 3:
        # Define the path to the folder containing the Excel files
        folder_path = r"//INN5FILER1V0001.corp.idemia.com/Work/MPB/Production/Data/MPB MIS/New DPR Report/2023/PastMonths/MayOnwardWithMachineDetails"

        # Create an empty list to store the Excel files
        excel_files = []

        # Loop through the directory and its sub-directories to find Excel files
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                if file.endswith(".xlsx")  and "DPR" in file:
                    excel_files.append(os.path.join(root, file))

        # Define the name of the worksheet you want to combine
        sheet_name = "Banking Perso performance"

        # Define the list of columns you want to include
        column_list = ["Date","Total Man Power (as per HR Record)","Total Present man Power",
                    "OT Operators","Total OT HRs","Total Man Hours"]

        # Create an empty DataFrame to store the combined data
        combined_df = pd.DataFrame()

        # Loop through each Excel file and read the specified worksheet into a DataFrame
        for file in excel_files:
            try:
                df = pd.read_excel(file, sheet_name=sheet_name, header=2)
                df = df[column_list] # select only specified column 
                combined_df = combined_df.append(df, ignore_index=True)

            except Exception as e:
                print(f"Error reading worksheet '{sheet_name}' from file '{file}': {e}")

            # specify the file name
        file_name = 'r"//INN5FILER1V0001.corp.idemia.com/Work/FI Data Analytics/02. Datasets/DPR/PP_MumbaiPM.xlsx"'     
            # check if the file already exists
        if os.path.exists(file_name):
            os.remove(file_name)  # delete the file if it exists

        combined_df.to_excel(r"//INN5FILER1V0001.corp.idemia.com/Work/FI Data Analytics/02. Datasets/DPR/PP_MumbaiPM.xlsx", index=False)
        
def PP_MumbaiCM():

    # Define the path to the folder containing the Excel files
    folder_path = r"//INN5FILER1V0001.corp.idemia.com/Work/MPB/Production/Data/MPB MIS/New DPR Report/2023/CurrentMonth"

    # Create an empty list to store the Excel files
    excel_files = []

    # Loop through the directory and its sub-directories to find Excel files
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".xlsx")  and "DPR" in file:
                excel_files.append(os.path.join(root, file))

    # Define the name of the worksheet you want to combine
    sheet_name = "Banking Perso performance"

    # Define the list of columns you want to include
    column_list = ["Date","Total Man Power (as per HR Record)","Total Present man Power",
                   "OT Operators","Total OT HRs","Total Man Hours"]

    # Create an empty DataFrame to store the combined data
    combined_df = pd.DataFrame()

    # Loop through each Excel file and read the specified worksheet into a DataFrame
    for file in excel_files:
        try:
            df = pd.read_excel(file, sheet_name=sheet_name, header=2)
            df = df[column_list] # select only specified column 
            combined_df = combined_df.append(df, ignore_index=True)

        except Exception as e:
            print(f"Error reading worksheet '{sheet_name}' from file '{file}': {e}")

        # specify the file name
    file_name = 'r"//INN5FILER1V0001.corp.idemia.com/Work/FI Data Analytics/02. Datasets/DPR/PP_MumbaiCM.xlsx"'     
        # check if the file already exists
    if os.path.exists(file_name):
        os.remove(file_name)  # delete the file if it exists

    combined_df.to_excel(r"//INN5FILER1V0001.corp.idemia.com/Work/FI Data Analytics/02. Datasets/DPR/PP_MumbaiCM.xlsx", index=False)
        
schedule.every().day.at("15:03").do(PP_MumbaiPM) 
schedule.every().day.at("11:01").do(PP_MumbaiCM)
schedule.every().day.at("15:03").do(PPNoidaPM) 
schedule.every().day.at("15:26").do(PPNoidaCM)
while True:
    schedule.run_pending()
    time.sleep(1)
```
